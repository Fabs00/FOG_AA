Aplica algún método de filtro a tus datos mediante el uso de SelectKBest

```python
from sklearn.feature_selection import SelectKBest, f_classif
```

```python
columnas =["ActiveCases","Serious,Critical",	"Tot Cases/1M pop",	"Deaths/1M pop",	"Tests/1M pop"]
selec_df = df[columnas]
selec = selec_df.fillna(selec_df.mean())
x = selec.values
columna =["TotalTests"]
selec_df1 = df[columna]
sele1 = selec_df1.fillna(selec_df1.mean())
y = sele1.values
```

```python
selector = SelectKBest(score_func=f_classif, k=5)
X_new = selector.fit_transform(x, y)
X_new
```
